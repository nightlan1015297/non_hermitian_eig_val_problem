```python
import numpy as np
from numpy import linalg as LA
from scipy.sparse.linalg import eigsh
from scipy import sparse
import plotly.graph_objects as go
```

```python
h_bar = 1
omega = 1
mass = 1
N_matrix = 10
alpha = 0.1

```

```python
def matrix_constructor(alpha):
    D_0 = np.array([[h_bar*omega*(n+1/2) if m==n else 0 for n in range(N_matrix)] for m in range(N_matrix)])
    D_1 = np.array([[np.sqrt(n+1)*alpha*1*np.sqrt(h_bar/(2*mass*omega)) if m==n+1 else 0 for m in range(N_matrix)] for n in range(N_matrix)])
    D_2 = np.array([[np.sqrt(n)  *alpha*1*np.sqrt(h_bar/(2*mass*omega)) if m==n-1 else 0 for m in range(N_matrix)] for n in range(N_matrix)])
    return D_0 + D_1 + D_2
```

```python
fig = go.Figure()

result_gnd = []
result_exci_1 = []

for N_matrix in range(2,500):
    
    

    H = matrix_constructor(alpha)
    val,vec = LA.eig(H)
    result_gnd.append(val[0])
    result_exci_1.append(val[1])



fig.add_trace(go.Scatter(x= [i for i in range(2,500)],y =np.array(result_gnd), name=f"GND"))
fig.add_trace(go.Scatter (x= [i for i in range(2,500)],y =np.array(result_exci_1) , name=f"1st Excited"))

fig.show()

```

```python
N_points = 1000
up_bound = 1
dn_bound = -1
alpha = 0

k = 1
m = 1

del_s = (up_bound - dn_bound)/(N_points - 1)

x_grid = np.linspace(dn_bound,up_bound,N_points,dtype=float)

D = sparse.spdiags(
    np.array([np.ones(N_points), -2*np.ones(N_points), np.ones(N_points)]),
    np.array([-1, 0,1]),
    N_points, N_points)

T_hat = (-1/2) * D /del_s**2

def potential (x_grid):
    return k*(x_grid)**2/2

def PT_symmetry_coefficient (x_grid):
    return 1*x_grid*alpha

V_hat = sparse.diags(potential(x_grid),(0))
U_hat = sparse.diags(PT_symmetry_coefficient(x_grid),(0))

H_hat = T_hat + V_hat + U_hat

E , psi = eigsh(H_hat,which='SM')

```
